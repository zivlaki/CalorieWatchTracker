```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = [
    ["M", 168, 22, 71, 38, 32, 37, 13.467, 40.400], 
    ["M", 185, 21, 69, 36, 33, 39, 14.829, 44.488],
    ["M", 184, 21, 70, 36, 30, 35, 14.749, 44.248],
    ["M", 230, 22, 73, 38, 44, 50, 18.437, 55.310],
    ["M", 143, 22, 65, 37, 28, 32, 11.463, 34.388],
    ["M", 148, 22, 69, 36, 26, 30, 11.863, 35.590],
    ["M", 200, 22, 75, 33, 36, 42, 16.032, 48.095],
    ["M", 145, 22, 74, 36, 25, 29, 11.623, 34.869],
    ["M", 171, 22, 69, 35, 36, 41, 13.707, 41.121]
        ]
metVal = 10.1
df = pd.DataFrame(data, columns=['Gender','Weight', 'Age','Height(Inches)', 'Calories Burnt By Treadmill','AW Active', 'AW Total', 'Calories Burnt By Formula', 'Total Burnt by Formula'])

print(df)
```

```python
awData = df['AW Active'].tolist()
treadData = df['Calories Burnt By Treadmill'].tolist()
formulaTotal = df['Total Burnt by Formula'].tolist()
weight = df['Weight'].tolist()


width1 = 0.3



print(awData)
print(treadData)
print(formulaTotal)
```

```python
# Line Plot 
plt.plot(awData, linestyle ="dotted", label="Apple Watch Active Calories Burnt")
plt.plot(treadData, linestyle = 'dotted', label="Calories Burnt By Treadmill")
plt.plot(formulaTotal, linestyle = 'dotted', label="Total Burnt By Formula")
plt.legend(fontsize = 8)
plt.show()
```

```python
# Generating x values for each group of bars
x = np.arange(len(weight))

# Creating the figure and subplot
fig, ax = plt.subplots()

# Plotting the first set of bars (AW Active)
bars1 = ax.bar(x - width1/2, awData, width1, label='AW Active')

# Plotting the second set of bars (Calories Burnt By Treadmill)
bars2 = ax.bar(x + width1/2, treadData, width1, label='Calories Burnt By Treadmill')

# Adding labels, title, and legend
ax.set_xlabel('Weight')
ax.set_ylabel('Value')
ax.set_title('Comparison of AW Active and Calories Burnt By Treadmill by Weight')
ax.set_xticks(x)
ax.set_xticklabels(weight)
ax.legend()

# Show the plot
plt.show()
```
